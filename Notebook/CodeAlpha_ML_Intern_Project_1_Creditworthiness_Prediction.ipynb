```python
#categorical columns in dataset 
object_cols = Credit_Worthiness.select_dtypes(include='object').columns
object_cols
```

```python
#insert LabelEncoder to convert categorical columns to numerical columns 
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
#loop for converting categorical columns to numerical columns 
for col in Credit_Worthiness : 
    if col in object_cols : 
        Credit_Worthiness[col] = encoder.fit_transform(Credit_Worthiness[col])
```

```python
#check if label column has imbalanced data 
Credit_Worthiness['Creditworthiness'].value_counts()
```

```python
#split dataset into features and label(Target column)
features = Credit_Worthiness.drop('Creditworthiness' , axis=1)
label = Credit_Worthiness['Creditworthiness']
```

```python
#the first 5 rows of features 
features.head()
```

```python
#the first 5 rows of label
label.head()
```

```python
#split features and label into(train , test)
from sklearn.model_selection import train_test_split
features_train , features_test , label_train , label_test = train_test_split(features , label , test_size=0.2 ,random_state=42)
```

```python
#insert RandomOverSampler to handle imbalanced data 
from imblearn.over_sampling import RandomOverSampler
Random_Sampler = RandomOverSampler()
```

```python
#handle imbalanced data using fit_resample
features_Random_Sampler , label_Random_Sampler = Random_Sampler.fit_resample(features ,label)
```

```python
#check that imbalanced data were handled 
label_Random_Sampler.value_counts()
```

```python
#split features and label after RandomOverSampler into train & test 
features_Random_Sampler_train , features_Random_Sampler_test , label_Random_Sampler_train , label_Random_Sampler_test = train_test_split(features_Random_Sampler ,label_Random_Sampler , test_size=0.2 , random_state=42)
```

```python
#number of rows and columns of features and label train 
print(features_Random_Sampler_train.shape)
print(label_Random_Sampler_train.shape)
```

```python
#number of rows and columns of features and label test
print(features_Random_Sampler_test.shape)
print(label_Random_Sampler_test.shape)
```

```python
#insert Standardscaler to scale the large range of colunms values 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
#scale the large range of colunms values after handle imbalanced data to avoid data leakage 
features_Random_Sampler_train[['Income' , 'Debt' , 'Loan_Amount' ,'Credit_Score']] = scaler.fit_transform(features_Random_Sampler_train[['Income' , 'Debt' , 'Loan_Amount' ,'Credit_Score']])
features_Random_Sampler_test[['Income' , 'Debt' , 'Loan_Amount' ,'Credit_Score']] = scaler.transform(features_Random_Sampler_test[['Income' , 'Debt' , 'Loan_Amount' ,'Credit_Score']])
```

```python
#import all classification models
from sklearn.linear_model import Perceptron , LogisticRegression
from sklearn.svm import SVC 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier , AdaBoostClassifier 
from sklearn.neighbors import KNeighborsClassifier 
from xgboost import XGBClassifier 
from catboost import CatBoostClassifier 
import lightgbm as lgb 
from sklearn.metrics import classification_report , accuracy_score , confusion_matrix
```

```python
#import all classification models we use in dictionary
models = {
    'Perceptron Model' : Perceptron(),
    'LogisticRegression Model' : LogisticRegression(),
    'SVC Model' : SVC() ,
    'DecisionTree Model' : DecisionTreeClassifier(),
    'RandomForest Model' : RandomForestClassifier(),
    'KNeighbors Model' : KNeighborsClassifier(),
    'AdaBoost Model' : AdaBoostClassifier(),
    'CatBoost Model' : CatBoostClassifier(),
    'XGB Model' : XGBClassifier(),
    'LightGBM Model' : lgb.LGBMClassifier()
    
}
```

```python
#visualize metrics of models
def visualize_matrix(actual , predicted) : 
    plt.Figure(figsize=(8,4))
    sns.heatmap(confusion_matrix(actual , predicted) , annot=True , fmt='d' )
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

```python
#Train the model, generate predictions, and evaluatethe model 
def impute_model(model) : 
    model.fit(features_Random_Sampler_train , label_Random_Sampler_train)
    train_pred = model.predict(features_Random_Sampler_train)
    test_pred = model.predict(features_Random_Sampler_test)

    print('Train classification report : \n',classification_report(label_Random_Sampler_train , train_pred))
    visualize_matrix(label_Random_Sampler_train , train_pred)

    print('Test classification report : \n', classification_report(label_Random_Sampler_test , test_pred)) 
    visualize_matrix(label_Random_Sampler_test , test_pred)

    print('==============================')
```

```python
#loop for all models 
for model_name , model_algo in models.items() : 
    print(f'================== {model_name} ==================')
    impute_model(model_algo)
    print('============================================')
```

```python
#Hyperparameter tuning with RandomizedSearchCV , GridSearchCV
from sklearn.model_selection import GridSearchCV , RandomizedSearchCV
```

```python
#Dict for all hyperparameters of models we try 
models_with_params = {
    'Decision Tree Model' : {
        'model' : DecisionTreeClassifier() ,
        'params' : {
            'criterion' : ['gini' ,'entropy' ,'log_loss'],
            'max_depth' : [None ,3,5,10,20],
            'min_samples_split' : [2,5,10],
            'min_samples_leaf' : [1,2,5,10]
        }
    },

    'Random Forest Model' : {
        'model' : RandomForestClassifier(),
        'params' : {
            'n_estimators' : [50 ,100 ,200 ,500] ,
            'max_depth' : [None ,3,5,10,20] ,
            'criterion' : ['gini' ,'entropy' ,'log_loss'],
            'min_samples_split' : [2,5,10],
            'max_features' : ['sqrt' ,'log2' ,None] 
        }
    },

     'catboost Model' : {
        'model' : CatBoostClassifier() ,
        'params' : {
            'iterations': [100, 300, 500],
            'depth': [4, 6, 8, 10],
            'learning_rate': [0.01, 0.05, 0.1],
        }
    },

    'XGBoost Model' : {
        'model' : XGBClassifier() ,
        'params' : {
            'n_estimators': [100, 200, 500],
            'max_depth': [3, 5, 7, 10],
            'learning_rate': [0.01, 0.05, 0.1],
            'subsample': [0.5, 0.7, 1],
            'colsample_bytree': [0.5, 0.7, 1],
        }
    },

    'LightGBM Model' :{
        'model' : lgb.LGBMClassifier() ,
        'params' : {
            'n_estimators': [100, 200, 500],
            'max_depth': [3, 5, 7, 10],
            'learning_rate': [0.01, 0.05, 0.1],
            'subsample': [0.5, 0.7, 1],
            'colsample_bytree': [0.5, 0.7, 1],
        }
    }

}
```

```python
#loop for all models with its parameters using RandomizedsearchCv and save model , best score , best params in list 
scores = []
for model_name , model_params in models_with_params.items():
    random_search = RandomizedSearchCV(model_params['model'] , model_params['params'] , cv=5 , n_iter= 30 , n_jobs=-1 , return_train_score=False , scoring='accuracy' )
    random_search.fit(features_Random_Sampler_train , label_Random_Sampler_train)

    scores.append({
        'model' : model_name ,
        'best_score' : random_search.best_score_ ,
        'best_params' : random_search.best_params_
    })

random_search_result = pd.DataFrame(scores , columns=['model' , 'best_score' , 'best_params'])
random_search_result
```

```python
#list of models and its params resulted from RandomizedSearchCV
scores
```

```python
#hyperparameter tuning with GridSearchCv 
model = RandomForestClassifier()

grid_params = {
   'n_estimators': [ 250 , 500 , 1000],
   'min_samples_split': [2.5 , 5 , 10],
   'max_features': ['log2'],
   'max_depth': [None],
   'criterion': ['log_loss']
}

grid_search = GridSearchCV(model , grid_params , n_jobs=-1 , cv=5 , scoring='accuracy' )
grid_search.fit(features_Random_Sampler_train , label_Random_Sampler_train)

print('Best Parameters : ' , grid_search.best_params_)
print("Best Score : " , grid_search.best_score_)
```

```python
#best model with its params resulted from gridsearchcv
model = RandomForestClassifier(criterion='log_loss' , max_depth=None , max_features='log2' , min_samples_split=5 ,n_estimators=1000) 
model.fit(features_Random_Sampler_train , label_Random_Sampler_train)
```
